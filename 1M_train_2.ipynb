```python
def renaming_dirs():
    dirs = next(os.walk(path))[1]
    counter = 0
    for d in dirs:
        os.chdir(path)
        crdir = os.getcwd()
        os.rename(d, list_folder_mids[counter])
        counter = counter + 1          
        
renaming_dirs()
```

```python
def renaming_data():
    dirs = next(os.walk(path))[1]
    for d in dirs:
        dirPath = path  + d
        os.chdir(dirPath)     
        files = next(os.walk(dirPath))[2]        
        counter = 0       
        for f in files:         
            img_path = os.path.join(dirPath, f) 
            os.rename(img_path,(os.path.join(dirPath, d + '_' + str(counter)+'.jpg')))
            counter = counter+1
            
renaming_data()
```

```python
def copying_data():
    dirs = next(os.walk(path))[1]
    for d in dirs:
        dirPath = path  + d
        os.chdir(dirPath)
        files = next(os.walk(dirPath))[2]
        for f in files:
            img_path = os.path.join(dirPath, f)
            shutil.copy2(img_path, TRAIN_DIR)
            
copying_data()
```

```python
dirs = next(os.walk(path))[1]

list_of_train_images = os.listdir(TRAIN_DIR)
list_of_train_images.remove('.DS_Store')

def creating_label(img):
    counter = 0
    for d in dirs:
        word_label = img.split('_')[0]
        if word_label == d:
            return counter
        counter = counter + 1
        
        
def creating_train_data():
    train_data = []
    for img in tqdm.tqdm(list_of_train_images):
        label = creating_label(img)
        img_path = os.path.join(TRAIN_DIR,img)
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMAGE_SIZE,IMAGE_SIZE), interpolation = cv2.INTER_AREA)
        train_data.append([np.array(img),np.array([label])])
    #os.chdir(dir_for_save)
    #np.save('train_data_zooqi_large_dtst.npy', train_data)
    shuffle(train_data)
    return train_data


train_data = creating_train_data() 
train_data
```

```python
train_X = [i[0] for i in train_data]
train_X = np.array(train_X).reshape(-1,IMAGE_SIZE,IMAGE_SIZE,1)
train_y = [i[1] for i in train_data]
#train_y = np.array(train_y).reshape(-1,1)
train_y
```

```python
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()

train_y = enc.fit_transform(train_y).todense()
print('Shape of data is: ',train_y.shape)
print('Type of data is: ',type(train_y))
train_y
```

```python
train_y[5]
```

```python
CNN = input_data(shape=[None, IMAGE_SIZE, IMAGE_SIZE, 1], name='input')

CNN = conv_2d(CNN, 32, 2, activation='relu')
CNN = max_pool_2d(CNN, 2)

CNN = conv_2d(CNN, 64, 2, activation='relu')
CNN = max_pool_2d(CNN, 2)

CNN = fully_connected(CNN, 1024, activation='relu')
CNN = dropout(CNN, 0.86)

CNN = fully_connected(CNN, 14, activation='linear')
CNN = regression(CNN, optimizer='adam', learning_rate=0.001, loss='categorical_crossentropy', name='targets')

model = tflearn.DNN(CNN)
model.fit({'input': train_X}, {'targets': train_y}, n_epoch=35, batch_size=128, snapshot_step=1000, show_metric=True)
```

```python
x = tflearn.input_data(shape=[None, 50, 50, 1], name='input')

x = conv_2d(x, 64, 3, activation='relu', scope='conv1_1')
x = conv_2d(x, 64, 3, activation='relu', scope='conv1_2')
x = max_pool_2d(x, 2, strides=2, name='maxpool1')

x = conv_2d(x, 128, 3, activation='relu', scope='conv2_1')
x = conv_2d(x, 128, 3, activation='relu', scope='conv2_2')
x = max_pool_2d(x, 2, strides=2, name='maxpool2')

x = conv_2d(x, 256, 3, activation='relu', scope='conv3_1')
x = conv_2d(x, 256, 3, activation='relu', scope='conv3_2')
x = conv_2d(x, 256, 3, activation='relu', scope='conv3_3')
x = max_pool_2d(x, 2, strides=2, name='maxpool3')

x = conv_2d(x, 512, 3, activation='relu', scope='conv4_1')
x = conv_2d(x, 512, 3, activation='relu', scope='conv4_2')
x = conv_2d(x, 512, 3, activation='relu', scope='conv4_3')
x = max_pool_2d(x, 2, strides=2, name='maxpool4')

x = conv_2d(x, 512, 3, activation='relu', scope='conv5_1')
x = conv_2d(x, 512, 3, activation='relu', scope='conv5_2')
x = conv_2d(x, 512, 3, activation='relu', scope='conv5_3')
x = max_pool_2d(x, 2, strides=2, name='maxpool5')

x = fully_connected(x, 4096, activation='relu', scope='fc6')
x = dropout(x, 0.5, name='dropout1')

x = fully_connected(x, 4096, activation='relu', scope='fc7')
x = dropout(x, 0.5, name='dropout2')

x = fully_connected(x, 14, activation='softmax', scope='fc8')

x = regression(x, optimizer='adam', learning_rate=0.001, loss='categorical_crossentropy', name='targets')

model = tflearn.DNN(x)
model.fit({'input': train_X}, {'targets': train_y}, n_epoch=10, batch_size=128, snapshot_step=1000, show_metric=True)
```
